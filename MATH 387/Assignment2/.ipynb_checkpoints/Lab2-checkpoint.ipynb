```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

$\texttt{randMatrix}$ generates a matrix $[a_{ij}] = A \in \mathbb{R}^{n \times n}$ such that $\forall i,j$,  $a_{ij}\sim U([-1,1])$

$\texttt{randNormalMatrix}$ generates a matrix $[a_{ij}] = A \in \mathbb{R}^{n \times n}$ such that $\forall i,j$,  $a_{ij}\sim \mathcal{N}(0,1)$

```python
# Function that generates a random matrix with entries 
def randMatrix(n):
    # np.random.rand generates an nxn matrix with random entries between 0 and 1.
    # We then scale this to be in the interval -1 and 1
    return np.matrix(np.random.rand(n,n)*2-1)

def randNormalMatrix(n):
    # np.random.rand generates an nxn matrix with random entries drawn from a standard normal distribution
    return np.matrix(np.random.randn(n,n))
```

```python
# Implemntation of Gaussian Elimination with Partial Pivoting
def GEPP(A):
    dims = A.shape
    n = dims[1]
    # Initialize row-interchange matrices
    diag = np.diagflat(np.ones(n))
    P = np.tile(diag,[n,1,1])
    L = np.tile(diag,[n,1,1])
    for i in range(n):
        # Find row with max element
        max_row = np.argmax(A[i:,i]) + i
        temp = diag[i,:]
        # Switch max row with row i 
        P[i,i,i] = 0
        P[i,i,max_row] = 1
        P[i,max_row,max_row] = 0
        P[i,max_row,i] = 1
        A = np.matrix(P[i,:,:])*A
        # Eliminate elements below pivot
        L[i,i+1:,i] = ( - A[i+1:,i] / A[i,i]).reshape(-1)
        A = np.matrix(L[i,:,:])*A
    return A

# Returns growth factor
def GF(A,U=False):
    if(type(U) is bool):
        U = GEPP(A)
    return np.max(abs(U))/np.max(abs(A))
```

```python
# This creates a scatter plot of matrix size against growth factor
def GF_Plot(randMatFcn,N1=10,N2=1010,R=50,step_count=20):
    D = np.empty([step_count,R])
    itr = range(N1,N2,int((N2-N1)/(step_count)))
    # Take R samples from step_count number of matrix sizes between N1 and N2
    for n in itr:
        for r in range(R):
            D[itr.index(n),r] = GF(randMatFcn(n))

    X = range(N1,N2,int((N2-N1)/step_count))
    # Calculate mean and standard deviation of growth factors
    M = np.mean(D,axis=1)
    S = np.std(D,axis=1)/np.sqrt(R)
    
    # Scatter plot
    fig = plt.figure(figsize=(10,5))
    plt.scatter(X,M)
    plt.ylabel('Growth Factor')
    plt.xlabel('n')
    plt.yscale('log')
    plt.xscale('log')

    # Calculate linear regression between log variables
    Y_log = np.log(M)
    X_log = np.log(X)
    A = np.vstack([X_log, np.ones(len(X_log))]).T
    m, c = np.linalg.lstsq(A, Y_log)[0]

    # Plot line of best fit
    plt.plot(X,X**m*np.exp(c),'-r')

    plt.legend(['Best Fit: alpha = %f' %m, 'Simulations'])

    plt.show()
    
    return D

def GF_dist(randMatFcn,N = [16,32,64,128]):
    # Calulate 5000 sample growth factors from matrices of sizes in N
    C = np.empty([4,5000])
    for n in N:
        for i in range(5000):
            C[N.index(n),i] = GF(randMatFcn(n))

    fig = plt.figure(figsize=(10,20))
    ax1 = fig.add_subplot(411)
    ax2 = fig.add_subplot(412)
    ax3 = fig.add_subplot(413)
    ax4 = fig.add_subplot(414)

    # Plot probility density histograms
    ax1.hist(C[0,:], bins=50, range=[0,2*N[0]],density=True)
    ax1.set_title('n = %d, n^(2/3)= %.2f, n^(1/2) = %.2f' % (N[0],N[0]**(2/3),N[0]**0.5))
    ax2.hist(C[1,:], bins=50, range=[0,N[1]*2], density=True)
    ax2.set_title('n = %d, n^(2/3)= %.2f, n^(1/2) = %.2f' % (N[1],N[1]**(2/3),N[1]**0.5))
    ax3.hist(C[2,:], bins=50, range=[0,N[2]*2],density=True)
    ax3.set_title('n = %d, n^(2/3)= %.2f, n^(1/2) = %.2f' % (N[2],N[2]**(2/3),N[2]**0.5))
    ax4.hist(C[3,:], bins=50, range=[0,N[3]*2],density=True)
    ax4.set_title('n = %d, n^(2/3)= %.2f, n^(1/2) = %.2f' % (N[3],N[3]**(2/3),N[3]**0.5))

    plt.show()
    
    return C
```

```python
# Calcualte the growth factor on 35 random matrices for 25 sizes evenly spaced between 10 and 1010
D1 = GF_Plot(randMatrix,10,1010,35,25)
```

Assuming that $y\approx cx^{\alpha}$ the slope of our regression line between the log of our variables gives us $\alpha$, which is close to 2/3. Clearly the trend of our growth factor is much better than $2^n$

```python
# Calculate PDF for uniform random matrices of sizes 10,20,40, and 80 from 5000 samples
C1 = GF_dist(randMatrix,[10,20,40,80])
```

The median of our distributions seem to be around $n^{2/3}$. The decay appears to be exponential.

```python
# Calcualte the growth factor on 5 random matrices for 10 sizes evenly spaced between 10 and 510 
# (ran out of time for computation)
D2 = GF_Plot(randNormalMatrix,10,1010,5,10)
```

This simulation once against clearly shows an a growth factor trend better than $2^n$, with an alpha similar to that from the uniform random case.

```python
# Calculate PDF for gaussian random matrices of sizes 10,20,40, and 80 from 5000 samples
C2 = GF_dist(randNormalMatrix,[10,20,40,80])
```

The average of our distributions appear to be slightly lower than in the uniform random case. We still have what appears to be expontential decay.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
from PIL import Image
```

The following piece of code defines the iteration map 

$$\phi(z)=z-\frac{z^3-2z-2}{3z^2-2}$$

of the Newton-Raphson method for the equation $z^3-2z-2=0$. In addition, the three roots are specified, and a color for each of the roots is chosen.

```python
#Newton-Raphson iteration for sin(z)=1
def phi(z):
    z=complex(z)
    return z-(z**3-2*z-2)/(3*z**2-2)
def f(z):
    return z**3-2*z-2
#Newton-Raphson iteration for sin(z)=1
def phi_stef(z,w):
    z=complex(z)
    N = (f(z+w*f(z)) - f(z))
    if(N==0):
        return np.inf
    else:
        return z - w*f(z)**2/N

zz = [np.sqrt(3),complex(-0.8846,0.5897),complex(-0.8846,-0.5897)]
cc=[[1,0,0],[0,1,0],[0,0,1]]
```

We consider the rectangle $(x_{\textrm{min}},x_{\textrm{max}})\times(y_{\textrm{min}},y_{\textrm{max}})$ in the $z$-plane, and superimpose on it a grid of size $\texttt{imgx}\times\texttt{imgy}$. From each grid point, we start the Newton-Raphson iteration, and depending on the root it converges to, we color the point. Convergence is deemed to occur if the cuurent iterand is within $\varepsilon$ distance of a root. In case the iteration takes too long (i.e., $\texttt{iteration number} > \texttt{maxit}$), or division by zero occurs, we leave the point uncolored (black). We also shade the colors in such a way that darker regions correspond to points that need more iterations to converge to a root.

```python
xmin = -.5
xmax = .5
ymin = -.5
ymax = .5
eps = .0001
maxit = 80
imgx = 500
imgy = 500
img = np.zeros((imgy,imgx,3))
a = (xmax-xmin)/imgx
b = (ymax-ymin)/imgy
h = np.zeros(maxit+1)
sit=20
sit_inv = 1./sit
for i in range(imgy):
    for j in range(imgx):
        x=xmin+a*j
        y=ymax-b*i
        z0=complex(x,y)
        for k in range(maxit):
            if abs(z0)<eps:
                h[maxit]+=1
                break
            z=phi(z0)
            con=False
            for l in range(len(zz)):
                if abs(z-zz[l])<eps:
                    con=True
                    break
            if(con):
                img[i,j]=[(maxit-k)*sit_inv*c for c in cc[l]]
                h[k]+=1
                break
            z0=z
        if k==maxit: h[maxit]+=1
plt.figure(figsize=(15,15))
plt.imshow(img,extent=[xmin,xmax,ymin,ymax])
plt.show()
scipy.misc.imsave('newt.png', img)
plt.figure(figsize=(14,7))
plt.plot(range(maxit+1),h)
plt.xlabel('# of iterations to convergence')
plt.ylabel('# of points')
plt.show()
```

```python
xmin = -.5
xmax = .5
ymin = -.5
ymax = .5
eps = .0001
maxit = 80
imgx = 500
imgy = 500
img = np.zeros((imgy,imgx,3))
a = (xmax-xmin)/imgx
b = (ymax-ymin)/imgy
h = np.zeros(maxit+1)
sit=20
sit_inv = 1./sit
omeg_list = [1,1.5,2]

fig = plt.figure(figsize=(10,30))
ax=[0,0,0]
ax[0] = fig.add_subplot(311)
ax[1] = fig.add_subplot(312)
ax[2] = fig.add_subplot(313)

for omeg in omeg_list:
    for i in range(imgy):
        for j in range(imgx):
            x=xmin+a*j
            y=ymax-b*i
            z0=complex(x,y)
            for k in range(maxit):
                if abs(z0)<eps:
                    h[maxit]+=1
                    break
                z=phi_stef(z0,omeg)
                con=False
                for l in range(len(zz)):
                    if abs(z-zz[l])<eps:
                        con=True
                        break
                if(con):
                    img[i,j]=[(maxit-k)*sit_inv*c for c in cc[l]]
                    h[k]+=1
                    break
                z0=z
            if k==maxit: h[maxit]+=1
    ax[omeg_list.index(omeg)].imshow(img,extent=[xmin,xmax,ymin,ymax])

plt.show()
scipy.misc.imsave('newt.png', img)
plt.figure(figsize=(14,7))
plt.plot(range(maxit+1),h)
plt.xlabel('# of iterations to convergence')
plt.ylabel('# of points')
plt.show()
```

Let $\alpha$ be a root of $f$, and assume $x\in\mathbb{R}$ in a neighbourhood of $\alpha$,
$$\phi'(x) = \frac{(- x^3 + 2x + 2)^2(3(3x^2 - 1)(- x^3 + x + 2)^2 - 9x^2 + 4)}{(4x - (- x^3 + x + 2)^3 - 3x^3 + 4)^2} + \frac{2(3x^2 - 2)(- x^3 + 2x + 2)}{4x - (- x^3 + x + 2)^3 - 3x^3 + 4} + 1$$
Hence, if $x\approx\alpha$, $x^2-2x-2\approx0$ and
$$\phi'(x)=1$$
Thus, by Taylor's Theorem, and the generalized mean value theorem,
$$ \Rightarrow |\phi(x)-\alpha| < \frac{|x-\alpha|^2}{2}f''(\xi)<|x-\alpha|^2M $$
